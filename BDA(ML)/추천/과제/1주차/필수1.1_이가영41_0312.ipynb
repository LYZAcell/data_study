# 필수과제 1.1
- event에 대한 유저들의 퍼널율 계산
- 각 이벤트 유형에 대한 사용자 수를 집계하고 퍼널 구성
1. 이벤트 유형에 따라 데이터 필터링
2. 이벤트 유형별 사용자 수 집계(event_type을 View, Cart, Purchase) 등으로 분류
3. 퍼널율 계산
  - 뷰율: 총 뷰 수
  - 장바구니율: 장바구니에 담은 수 / 뷰 수
  - 구매율: 구매 수 / 장바구니 수

```python
def calculate_funnel_rates(df):
    # 각 이벤트 유형별 고유 사용자 수 집계 _ nunique = 독특한 값 찾
    view_count = df[df['event_type'] == 'view']['user_id'].nunique()
    cart_count = df[df['event_type'] == 'cart']['user_id'].nunique()
    purchase_count = df[df['event_type'] == 'purchase']['user_id'].nunique()

    # 퍼널율 계산
    cart_conversion_rate = cart_count / view_count if view_count > 0 else 0
    purchase_conversion_rate = purchase_count / cart_count if cart_count > 0 else 0

    return {
        '뷰-카트 전환율': cart_conversion_rate,
        '카트-구매 전환율': purchase_conversion_rate
    }

# 각 월별 퍼널율 계산
oct_funnel_rates = calculate_funnel_rates(df_oct)
nov_funnel_rates = calculate_funnel_rates(df_nov)
dec_funnel_rates = calculate_funnel_rates(df_dec)

# 결과 출력
print("10월 퍼널율:", oct_funnel_rates)
print("11월 퍼널율:", nov_funnel_rates)
print("12월 퍼널율:", dec_funnel_rates)
```
