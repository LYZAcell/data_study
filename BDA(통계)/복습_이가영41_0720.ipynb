```python
## 표본 추출방법
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# 데이서셋 생성
np.random.seed(10)
data = pd.DataFrame({
    'Feature1': np.random.randn(100),
    'Feature2': np.random.randn(100),
    'Group': np.random.choice(['A','B','C','D'],100)
})
```

```python
# 무작위 추출 simple random sapmpling
data.sample(n=20, random_state =1)
```

```python
# 층화 추출
data.groupby("Group", group_keys = False).apply(lambda x: x.sample(frac=0.5))
```

```python
# 군집 추출 cluster sampling
# 특정 군집을 지정해서 추출
cls = data['Group'].unique()
selected_cls = np.random.choice(cls)
cluster_sample = data[data['Group'] == selected_cls]
print(cluster_sample)
```

```python
# 계통 추출
k = 5
systematic_sample_data = data.iloc[::k,:]
print(systematic_sample_data)
```

# 타이타닉 자료로 분포 확인하기

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
tt = sns.load_dataset('titanic')
titanic = tt.dropna(subset = 'age')
titanic
```

- 타이타닉 데이터에서 샘플표본을 추출 -> 추정량 계산 -> 신뢰구간 연산 -> 전체데이터 모집단 비교
- 표본에 대한 값으로 모집단 age 를 추출해보기!

```python
# 표본 크기
sample_size = 100
np.random.seed(1)

# 무작위 추출
sp = titanic['age'].sample(n=sample_size, random_state = 1)

# 추정량 계산 _ 평균과 표준오차
sample_mean = sp.mean()
sample_std = sp.std(ddof = 1) # 샘플의 표준편차
```

```python
# 표준오차
std_error = sample_std/np.sqrt(sample_size)
std_error
```

```python
# 신뢰구간(95%구간)
cof_level = 0.95
degrees_freedom = sample_size -1

# 신뢰구간 stats 쉽게 구할 수 있다
cof_interval = stats.t.interval(cof_level, degrees_freedom, sample_mean, std_error) # 신뢰구간 level, 자유도, 샘플 평균 샘플 표준오차
cof_interval # 25살 ~ 32살이 95% 신뢰할만한 구간이다
```

```python
# 결과 출력해보기
print(f'표본 크기: {sample_size}')
print(f'표본 평균: { sample_mean: .2f}')
print(f'표본 표준편차: {sample_std: .2f}')
print(f'표준오차: {std_error: .2f}')
print(f'신뢰구간: {cof_interval}')
```

```python
# 전체 titanic 데이터 비교
tt_mean = titanic['age'].mean()
print(f'모집단 평균: {tt_mean:.2f}')
```

```python
# 시각화를 통해 그래프 그려보기 : 모집단의 분포를 알아보기
# 모집단과 표본집단의 평균이 어느정도 일치하는지 확인하는 방법
plt.figure(figsize = (10,7))
sns.histplot(titanic['age'], kde = True, color ='blue', label = 'tt')
sns.histplot(sp, kde = True, color = 'red', label = 'Sample', alpha = 0.5)
plt.axvline(sample_mean, color = 'red', linestyle = '--', label = 'Sammple Mean')
plt.axvline(tt_mean, color = 'blue', linestyle = '--', label = 'TT Mean')
plt.axvline(cof_interval[0], color = 'green', linestyle = '--', label = 'Confidence Interval Lower Bound')
plt.axvline(cof_interval[1], color = 'green', linestyle = '--', label = 'Confidence Interval Upper Bound')
plt.legend()
plt.show()
```

# 신뢰구간

## 신뢰구간 : (점추정치 - 오차한계) <= 모비율(모평균) >= (점추정치 + 오차한계)

- 방금 우리는 나이에 대한 신뢰구간 추정을 해봤다.
- 신뢰 추정 모수는 얼마든지 달라질 수 있기에 고민하는 과정이 필요하다

## 신뢰한계
- 점추정치 +- 오차한계
- 점추정치 +- 표준오차(표준편차)
* 표준오차는 1/루트n에 비례함


## 모비율의 이해
- 흡연율을 알기위해 500명을 무작위로 추출하여 흡연 유무 조사 -> 150명이 흡연 중
- 전교생 흡연율에 대한 95%신뢰 구간 추정
- 점추정: 150/500 = 0.3 (30%)
- 구간추정 (신뢰한계 ) = 0.3 +- 1.06(루트)(0.3(1-0.3)/500) = 0.3 +- 0.04

- 샘플에 bias가 끼면 깨지기 너무 쉬운 수치기 때문에 통계 장난질에 대한 견제가 필요함

## 모평균의 이해
- 표본 추출- > 표본평균(추정량) -> 점추정 / 구간추정하며 신뢰한계 정하기
- 아웃라이어가 나오면 신뢰할 수 없는 구간이라고 가설점정하는 과정이 필요


### 모집단을 모른다고 생각하고 표본추출로 연산하기
1. 모표준편차를 아는 경우: x +- Z(a/2)* 모표준편차 / 루트n
2. 모르는 경우  - 표본이 큰 경우 (30개 이상) _ 동일
3. 모르는 경우 _ 표본이 적은 경우 : x +-ta/2(^n-1)~ 동일
