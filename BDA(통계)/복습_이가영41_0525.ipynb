# 이산형 확률변수

**확률, 확률변수를 넘어가고 확률 분포로 당도할 것**

## 확률분포

- 확률: 주사위를 던졌을 때 3이 나올 확률 1/6
- 확률 변수: 주사위에서 나올 수 있는 숫자(1~6)
  - 확률 변수는 이산형, 연속형 모두 가능해진다
  - 이산형: 정수형 (가족수, 불량품 수 등)
  - 연속형 변수: 연속적 값. 한 점에서 정의할 수 없는. (키, 체중, 임금, 통금시간 등)
- 확률 분포: 주사위를 던졌을 때 X(확률변수)가 각각 값을 가질 확률 (P(X=1) = 1/6 )

- 우리는 이산 확률분포에서 베르누이, 이항분포까지 갈 것


### 베르누이 시행


- 두가지 상호 배타적인 결과만 나오는실험 ( 동전던지기, 양성/음성 등)
- 결과는 보통 0이나 1

#### 이항분포
- 베르누이가 n번 반복될 때 각 베르누이 실험이 독립적이고 각 실험에서의 파이는 일정하게 유지된다. 이항확률 변수 X는 n개의 독립적인 베르누이 확률 변수의 합 (X = X1+X2+...Xn)
- 이항분포의 평균: n개의 베르누이 평균의 합, 분산 확인 (n파이, n파이(1-파이)

- 이항분포의 모양은 파이의 값에 따라 달라짐
- 복원추출의 성질을 갖고 있고 , P = 0.5일 때 기대값 E(X) = np에 대해 좌우 대칭을 띈다
- np >= 5이고 np(1-p) >= 5일 때 정규분포로 근사
- P <=0.1이고 np =0.1 ~10일 때 포아송 분포로 근사

# 이항분포 코드 만들기
- scipy활용

```python
#함수 호출
import numpy as np
import matplotlib.pyplot as plt
# scipy에서 베르누이 불러오기
from scipy.stats import bernoulli
```

## 베르누이 분포

- from scipy.stats import bernoulli
- 매소드 : 확률질량함수 pmf(k,p): p 특정확률 k가 0또는 1( 성공, 실패)
- 누적분포 함수 cdf
- 랜덤변수 샘플링 rvs(p,size)p는 특정확률 size. 사이즈 크기 성공확률이 p인 랜덤변수 생성가능
- mean(p)확률
- var(p), std(p)

```python
p = 0.5

# 성공 실패 이진
x = [0, 1]
pmf = bernoulli.pmf(x, p)
plt.bar(x,pmf,width = 0.3)
```

```python
# cdf 계산
cdf = bernoulli.cdf(x,p)
plt.step(x,cdf)
```

```python
# 랜덤 샘플링 시각화
rvs = bernoulli.rvs(p, size = 1000)
bins = [-0.5,0.5,1.5]
plt.hist(rvs,bins=bins)
```

## 이항분포 (binom)

- 이항분포를 위해 베르누이를 n번 시행
- 이를 binom으로 구현 가능

- binom
  - pmf(k,n,p) p확률, n특정 시행횟수, k번 성공할 확률
  - cdf(k,n,p)
  - sf(k, n, p) 생존함수
  - ppdf(q,n.p) 퍼센트 포인트 함수, 역누적분포함수, 누적확률 q에 대해서 해당하는 이항분포 성공횟수 k의 계산
  - rvs(n,p,size): 랜덤변수 샘플링
  - 기타 통계치 동일

```python
from scipy.stats import binom
```

```python
n = 10 # 시행횟수
p = 0.5 # 성공확률

# 랜덤샘플 시각화
rvs = binom.rvs(n,p,size = 1000)
# 시행횟수에 대한 수치를 쭉 작성해줌
rvs
```

```python
# 실제 데이터셋분석은? _ 생존률, 이탈률 예측 분석
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
# 유방암 관련 자료 로드
data = load_breast_cancer()
```

```python
# 특성 데이터와 타겟 데이터를 데이터프레임으로 변환
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# 데이터프레임 정보 출력
df.head(3)
```

```python
# 베르누이 분포라 가정하여 시뮬레이션을 돌리면 같값이 나올까?
X = data.data
y = data.target
len(y)
```

```python
p = np.mean(y)
simulated_data = bernoulli.rvs(p, size = len(y))
simulated_data
```

```python
# 실제 데이터와 비교
plt.figure(figsize = (10,6))
plt.hist(y, bins = 2, alpha = 0.3, color = 'b', label = 'Actual result data')
plt.hist(simulated_data, bins = 2, alpha = 0.3, color = 'r', label = 'Simulated data')
plt.legend()
plt.show()
```

- 가상의 고객이 이탈에 대해서 이탈을 샘플링
- 가상의 고객 데이터를 만들어야 한다
- 고객의 베르누이분포를 통해 이탈율에 대한 시각화

```python
from sklearn.datasets import make_classification # 이진데이터를 쉽게 만들 수 있다
X,y = make_classification(n_samples = 1000, n_features = 10, n_classes = 2, random_state = 111)

# 이탈확률 예측
p = np.mean(y)
simulated_data = bernoulli.rvs(p, size = len(y))
```

```python
# 실제 데이터와 비교
plt.figure(figsize = (10,6))
plt.hist(y, bins = 2, alpha = 0.3, color = 'b', label = 'Actual result data')
plt.hist(simulated_data, bins = 2, alpha = 0.3, color = 'r', label = 'Simulated data')
plt.legend()
plt.show()
```

- 이항분포가 시행횟수를 늘리면 -> 정규분포화 처럼 된다. np >= 5 이상이면

```python
# 이항분포 설정

n_trials = 50 #시행횟수
p = 0.5 #성공확률
n_simulations = 1000 #샘플크기

binom_samples = binom.rvs(n = n_trials, p = p, size = n_simulations)
#binom_samples
```

```python
# 시각화로 확인
plt.figure(figsize = (10,6))
plt.hist(binom_samples, bins = np.arange(n_trials + 2 ) - 0.5, density = True, alpha = 0.4, color = 'b')
plt.title('Binominal Distribution(n,p)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 타이타닉으로 확인
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
survived = df['survived'].dropna() # 결측치 제거
p = np.mean(survived)
p
```

```python
# 베르누이 생성
simulated_data = bernoulli.rvs(p, size = len(survived))
simulated_data
```

```python
actual_counts = survived.value_counts(normalize=True)
simulated_counts = pd.Series(simulated_data).value_counts(normalize=True)

# 막대 그래프 시각화
fig, ax = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# 실제 데이터 시각화
actual_counts.plot(kind='bar', ax=ax[0], color=['blue', 'green'], alpha=0.7)
ax[0].set_title('Actual Survival Distribution')
ax[0].set_xlabel('Survived')
ax[0].set_ylabel('Frequency')

# 시뮬레이션 데이터 시각화
simulated_counts.plot(kind='bar', ax=ax[1], color=['blue', 'green'], alpha=0.7)
ax[1].set_title('Simulated Survival Distribution')
ax[1].set_xlabel('Survived')

plt.show()
```
