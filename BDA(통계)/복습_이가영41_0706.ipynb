# 카이제곱 분포


- 표준정규분포 Z를 다 제곱해서 진행
- 적합도 검정, 독립성 검정 등 진행 시 사용. 연속, 이산
- 연속 -> 범주형 데이터셋을 분석하기 위한 하나의 분포
- 모분산에 대한 가설검정에 사용

### 심슨의 패러독스
- 교차표를 분석 시 각각의 부분자료에 성립하는 대소관계가 그 부분의 자료를 종합한 전체자료에서 성립하지 않는 경우

### 카이제곱분포 X^2의 특징
1. 연속형 확률분포로서 확률변수는 항상 양의 값만을 갖는다
2. 오른쪽 꼬리가 있는 비대칭 분포
3. 자유도, 즉 표준정규확률변수 Z제곱한 것을 몇개 더했나에 따라 모양이 결정된다. Z^2가5개라면 자유도는 5
  - 자유도: 독립적인 관측값의 갯수
    - 표본자료로부터 평균까지의 편차합은 항상 0
    - n-1개 자료의 편차를 알면 나머지 하나의 편차는 다른 자료들에 의해 자동 결정
    - 전체 n개의 자료값 중에서 n-1개의 자료값만 자유롭게 변화할 수 있다는 의미
4. 확률 변수 X가 자유도 v인 카이제곱의 분포를 따를 경우 X ~ X^2(v)
5. X~X^2(n)일 때 E(X) = n, Var(X) =2n

# t분포

- 확률변수 Z는 표준정규분포를, X^2은 자유도가 r인 카이제곱 분포를 따르며 서로 독립이면 즉 Z~N(0,1)rhk x^2 ~x^2(v)일 때 두 확률변수를 결합한 새로운 확률변수는 자유도가 r인 t분포를 한다

- **정규분포와 유사해보이지만 표본크기가 작고, 모분산이 알려져있지 않을 때 사용하는 확률분포**


### 즉 소표본을 다루는 통계 분포다
- 소표본에 대한 평균에 대한 신뢰구간 측정, 가설 검정 진행
- 자유도가 커질수록 t분포는 표준정규분포와 가까워진다.

# F분포

- 두 확률변수 x1^2과 x2^2이 자유도가 각각 v1,v2인 카이제곱 분포를 따르고, 서로 독립이면 즉 Z~N(0,1)과 x^2 ~ x^2(v)일 때 두 확률변수의 비로 나타낸 새로운 확률변수는 자유도가 분자, 분모 순으로 v1, v2인 분포

- F분포의 경우는 두개의 독립적인 카이제곱 분포를 따르는 변수의 비율로 정의되는 확률 분포. 주로 분산 분석에서 많이 사용(ANOVA). 두 집단간의 분산 비교를 위해 사용

### F분포의 의미
- 여러 집단의 평균을 분석하거나 그룹간의 분산과 그룹내의 분산을 비교할 때 그룹간의 차이가 유의미한지 검정
- 분산 비교: 두집단의 분산이 같은지 비교할 대 사용
  - i.e: 변동성이 동일한지 검정
  - 제품 공장 A,B가 있다면 A와 B의 제품의 품질의 변동성을 비교한다면 -> 각 공장에서 생산된 제품의 품질 데이터를 수집 > F검정을 사용해서 두 공정의 품질변동성이 동일한지 검정
  - 시험들의 성적 분포가 유의미한지 검증할 때

1. 연속형 확률분포로서 확률변수 F는 항상 양의 값만을 가진다
2. 오른쪽 꼬리를 가진 비대칭분포
3. 분자의 자유도와 분모의 자유도에 따라 분포의 모양이 변한다.
4. F분포의 면적은 자유도가 변하더라도 항상 1

# 추정과 가설 검정


1. 추정: 모수에 대해 정보가 거의 없는 상황에서 모수(모비율, 모평균, 모분산)등의 값은 얼마인가?에 대한 답을 표본을 추출한 정보를 이용해 통계적으로 답하는 것
2. 모집단에 대해 약간의 정보를 갖고 있지만 확실한지를 더 검증하기 위해 귀무가설/대립가설을 결정하고 채택여부를 결정하는 방법

### 전체벌점 추정
- 데이터 입문 통계반의 평균을 갖고 벌점 추정
- 우리반 평균벌점 10점 -> 추정 값 계산 -> 벌점 15점 연산 -> 1000명에 적용.,신뢰구간 10~20점 사이의 벌점이구나 연산

1. 추정량 : p햇 = X/n _ 추정량은 표본비율
2. 추정치: 추정치는 점추정과 구간추정으로 나뉜다
3. 점추정 : 추정하고자 하는 모수를 하나의 값으로 추정하는 방법
  - 단 추정할 모수에 대해  한개의 값으로 추정하니 추정량의 값이 모수로부터 얼마나 떨어져있는지에 대한 정보는 제공할 수 없음
4. 구간추정: 신뢰수준 추정

- 샘플링 시 분포와 특성을 고려해야한다


### 점추정 조건
1. 불편성: 추정량의 기댓값이 추정할 모수와 일치하는 성질
2. 유효성: 불편성을 만족시키면서 분산이 최소인 성질
3. 일치성: 표본의 크기가 증가할수록 추정량을 보다 신뢰할 수 있다는 의미
4. 충족성: 추정량이 모수에 관한 정보를 충분히 제공할 수 있는 정도. 이러한 조건을 가진 추정량을 충족 추정량


# 코드

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, chi2, f, t
```

```python
# 정규 분포
mu, sigma = 0,1
x_norm = np.linspace(-5,5,100)
y_norm = norm.pdf(x_norm, mu, sigma)
plt.plot(x_norm, y_norm)
plt.show()
```

```python
# 카이제곱 분포와 자유도
df_chi2 = 3
x_chi2 = np.linspace(0,10,100)
y_chi2 = chi2.pdf(x_chi2, df_chi2)
plt.plot(x_chi2, y_chi2)
plt.show()
```

```python
# F분포 : 자유도 두개를 비교하는 것
d1,d2 = 5,3
x_f = np.linspace(0,5,1000)
y_f = f.pdf(x_f, d1, d2)
plt.plot(x_f, y_f)
plt.show()
```

```python
# t분포
df_t = 10
x_t = np.linspace(-5,5,1000)
y_t = t.pdf(x_t, df_t)
```

```python
#그래프 그리기
plt.figure(figsize=(12,12))

# 정규분포
plt.subplot(4,1,1)
plt.plot(x_norm, y_norm, label='Normal')
plt.fill_between(x_norm, y_norm, alpha=0.2)
plt.legend

# 카이제곱분포
plt.subplot(4,1,2)
plt.plot(x_chi2, y_chi2, label='chi2')
plt.fill_between(x_chi2, y_chi2, alpha=0.2)
plt.legend

# F분포
plt.subplot(4,1,3)
plt.plot(x_f, y_f, label='F')
plt.fill_between(x_f, y_f, alpha=0.2)
plt.legend

# t분포
plt.subplot(4,1,4)
plt.plot(x_t, y_t, label='t')
plt.fill_between(x_t, y_t, alpha=0.2)
plt.legend
```
