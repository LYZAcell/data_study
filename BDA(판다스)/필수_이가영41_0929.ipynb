# Groupby

```python
mpg_ori = mpg.groupby('origin')
for i,j in mpg_ori:
  print(f'######################## origin : {i} ##########################')
  print(j)
```

```python
# get_group : groupby기준 분류 호출
mpg_ori.get_group('europe')
```

```python
# 여러개 기준 groupby가능
mpg_oricyl = mpg.groupby(['origin', 'cylinders'])
for i,j in mpg_oricyl:
  print(f'######################## origin,cylinders : {i} ##########################')
  print(j)
```

```python
mpg_oricyl.get_group(('usa', 6))
```

## agg
- 함수적용해서 특정한 통계치 보기

```python
# 문법 : agg(보고싶은거 = ('해당변수', '통계치'))
mpg_oricyl.agg(model_year_mean = ('model_year', 'mean'))
```

```python
##특정 함수를 적용
def max_ave(x):
    return x.max()- x.mean()

mpg_oricyl.agg(weight_max_ave = ('weight', max_ave))
```

```python
# groupby 응용
mpg.groupby('origin')['cylinders'].mean()
```

# qcut, cut
- qcut: 분위컷
- cut: 고정값으로 나누는 함수

```python
# 4분위수 : 개수기준
mpg_qcut = pd.qcut(mpg['horsepower'], q = 4, labels = ['1st','2nd','3th','4th'])
mpg_qcut
```

```python
# 일정한 크기의 구간을 정하고 구간해당. 숫자범위임
mpg_cut = pd.cut(mpg['horsepower'], bins = 4, labels = ['낮음','중간','높음','매우높음'])
mpg_cut
```

# 기타: apply, assign

```python
# assign을 이용해 새로운 열 km_per_liter 추가
mpg = mpg.assign(km_per_liter = lambda x: x['mpg'] * 0.425144)

print(mpg.head())
```

```python
# apply를 사용해 name 컬럼에서 제조사 추출
mpg['manufacturer'] = mpg['name'].apply(lambda x: x.split()[0])

print(mpg[['name', 'manufacturer']].head())
```

# 병합

```python
mpg_4 = mpg.head(4)
manufacturer_info = pd.DataFrame({
    'manufacturer': ['ford', 'chevrolet', 'toyota', 'bmw'],
    'headquarters': ['USA', 'USA', 'Japan', 'Germany'],
    'avg_price': [30000, 35000, 25000, 45000]
})

merged = pd.merge(mpg, manufacturer_info, on='manufacturer', how='left')
merged.head()
```

```python
# mpg 데이터 불러오기
mpg = sns.load_dataset('mpg')

# 제조사별 평균 연비 계산
mpg_avg = mpg.groupby('origin')[['mpg']].mean().reset_index()

new_data = pd.DataFrame({
    'manufacturer': ['ford', 'chevrolet', 'toyota', 'honda'],
    'new_data': [22.5, 20.1, 33.4, 35.0]  # 추가적인 연비 관련 데이터라고 가정
})

# 행 방향으로 concat
result = pd.concat([mpg_avg, new_data], axis=0, ignore_index=True)
result
```
