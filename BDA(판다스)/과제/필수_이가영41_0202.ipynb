# 필수과제1.
- 긍정과 부정에 주요 핵심 키워드를 카운팅 하여서 어떤 키워드가 가장 높은 빈도를 차지하는지?
- 긍정과 부정 두 개 비교해서 상위 5개 리스트 공유

```python
import pandas as pd
from collections import Counter
import re

df = pd.read_csv('./movie_rv.csv')
df.head()
```

```python
# 텍스트 전처리 함수
def preprocess_text(text):
    # Check if text is a string before processing
    if isinstance(text, str):
        text = re.sub(r'[^\w\s]', '', text)  # 특수문자 제거
        return text.split()  # 단어 리스트로 변환
    else:
        # Handle non-string values, e.g., return an empty list
        return []

# 긍정과 부정 리뷰 분리
positive_reviews = df[df['label'] == 1]['document']
negative_reviews = df[df['label'] == 0]['document']

# 긍정 키워드 카운트
positive_words = Counter()
for review in positive_reviews:
    words = preprocess_text(review)
    positive_words.update(words)

# 부정 키워드 카운트
negative_words = Counter()
for review in negative_reviews:
    words = preprocess_text(review)
    negative_words.update(words)

# 상위 5개 키워드 추출
top_positive = positive_words.most_common(5)
top_negative = negative_words.most_common(5)

# 결과 출력
print("긍정 키워드 상위 5개:", top_positive)
print("부정 키워드 상위 5개:", top_negative)
```
