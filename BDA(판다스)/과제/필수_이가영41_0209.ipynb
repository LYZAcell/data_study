```python
import pandas as pd
from konlpy.tag import Okt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import re
okt = Okt()
```

```python
df = pd.read_csv('movie_rv.csv')
df
```

```python
# NaN 값을 빈 문자열로 대체
df['document'] = df['document'].fillna('')

# 형태소 분석 및 불용어 제거 (입력이 문자열인지 확인)
def preprocess(text):
    if not isinstance(text, str):  # 입력이 문자열이 아닐 경우
        return ''

    # 숫자 및 특수 문자 제거
    text = re.sub(r'[^가-힣\s]', '', text)  # 한글과 공백을 제외한 모든 문자 제거

    stopwords = ['영화', '재미', '짜증', '목소리', '초딩', '연기', '가볍', '솔직히', '조정']
    tokens = okt.morphs(text)
    return ' '.join([word for word in tokens if word not in stopwords])

df['processed_document'] = df['document'].apply(preprocess)
```

```python
count_vectorizer = CountVectorizer()
count_vector = count_vectorizer.fit_transform(df['processed_document'])

# 차원 확인
print("Count Vector Shape:", count_vector.shape)
print("Count Vector Features:", count_vectorizer.get_feature_names_out())
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_vector = tfidf_vectorizer.fit_transform(df['processed_document'])

# 차원 확인
print("TF-IDF Vector Shape:", tfidf_vector.shape)
print("TF-IDF Vector Features:", tfidf_vectorizer.get_feature_names_out())
```

```python
# Count Vectorizer의 단어 목록
print("Count Vectorizer Features:", count_vectorizer.get_feature_names_out())

# TF-IDF Vectorizer의 단어 목록
print("TF-IDF Vectorizer Features:", tfidf_vectorizer.get_feature_names_out())
```
