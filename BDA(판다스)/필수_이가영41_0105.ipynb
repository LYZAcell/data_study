- 필수과제 1.2
    - name 별 groupby 데이터 요약집계 후 피처들의 관계를 시각화로 그려주세요! (가공후 데이터)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mpg_data = sns.load_dataset('mpg')
mpg_data.head()
```

```python
# 1.1 , 가공전 데이터 시각화
# 2x2 그래프 그리기 ( 가공전)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 산점도: 마력 대비 연비 (horsepower vs mpg)
sns.scatterplot(data=mpg_data, x='horsepower', y='mpg', ax=axes[0, 0])
axes[0, 0].set_title('Horsepower vs MPG')

# 박스플롯: 생산지에 따른 연비
sns.boxplot(data=mpg_data, x='origin', y='mpg', ax=axes[0, 1])
axes[0, 1].set_title('MPG by Origin')

# 히스토그램: 연비 분포도
sns.histplot(data=mpg_data, x='mpg', bins=20, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of MPG')

# 산점도: 무게 대비 연
sns.scatterplot(data=mpg_data, x='weight', y='mpg', ax=axes[1, 1])
axes[1, 1].set_title('Weight vs MPG')

plt.tight_layout()
plt.show()
```

```python
# 2x2 그래프 설정
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 6))

# 그래프 설정을 위한 리스트
plots = [
    {'func': sns.barplot, 'x': 'origin', 'y': 'mpg', 'title': 'Average MPG by Origin', 'ax': axs[0, 0]}, # 생산국에 따른 연비 평균
    {'func': sns.barplot, 'x': 'cylinders', 'y': 'mpg', 'title': 'Average MPG by Cylinders', 'ax': axs[0, 1]},#실린더 개수에 따른 연비 평균
    {'func': sns.histplot, 'x': 'mpg', 'title': 'Distribution of MPG', 'ax': axs[1, 0]}, # 연비 분포
    {'func': sns.histplot, 'x': 'horsepower', 'title': 'Distribution of Horsepower', 'ax': axs[1, 1]} #마력분포
]

# 그래프 그리기
for plot in plots:
    plot['func'](data=mpg_data, x=plot.get('x'), y=plot.get('y', None), ax=plot['ax'])
    plot['ax'].set_title(plot['title'])

plt.tight_layout()  # 겹치는 것 방지
plt.show()
```
