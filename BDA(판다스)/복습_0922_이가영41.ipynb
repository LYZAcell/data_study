## 파생변수 만들기
- 기존변수를 이용해 새로운 변수를 만들기
- 기존의 변수의 영향을 받기 때문에 정말 꼭 만들어야하는지에 대한 고민 후 ML등의 피처로 사용할 때도 기존변수와의 관계도 고려해야한다
- 무조건 파생변수를 많이 만드는건 차원이 더 추가되는 것과 같아서 조심해야한다
- 원래의 변수들을 가공하여 만드는 것
```
bda_class['파생변수'] = bda_class['원래변수'] - bda_class['원래변수2']
```
- 변수처럼 사용할 수 있다

````python
bda_class=pd.DataFrame({'이름':['김길동','이길동','박길동','최길동','홍길동'],
        '상점':[10,20,30,40,50],
             '벌점':[1,2,3,5,8]})
````

````python
# 넘파이로 인덱스 접근 가능
np.where(bda_class['최종상태'] >= 20)

# 파생변수 만들기
bda_class['수료여부'] = np.where(bda_class['최종상태']>= 20, 'pass','fail')
````

````python
bda_class['수료여부'].value_counts()
````
