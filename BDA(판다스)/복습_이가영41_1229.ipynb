## 데이터 병합 (Merge)

조건에 따라 아래의 함수를 선택하여 쓰면 된다
- 다만 잘못된 방법으로 병합할 시 문제 발생
- 병합은 필수적인 요소. 머신러닝 분야는 필수로 병합과정이 들어가고 분석에서도 필수적인 분야
- 원하는 데이터셋을 구성하기 위해서는 필수적이다.

## 병합 함수들
- concat
- merge
- join

## 코드

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
df2 = pd.DataFrame({'A':[4,5],'B':[6,7]})
df3 = pd.DataFrame({'A':[11,12],'B':[6,7],'C':[100,200]})
```

### Concat:  덩어리 + 덩어리 ( 데이터 프레임 + 데이터프레임)
데이터프레임

+

데이터프레임 형식으로 붙는다

- 축 지정이 필요하다. axis = 0 (행) or 1 (열 기준)
- concat 후 N/A값과 index에 유의 해야한다 ( 기존의 인덱스를 유지하여 0,1,2, 0-1을 유지할 수도 )

```python
display(df1, df2, df3)
```

```python
# df1 + df2
pd. concat([df1, df2], axis = 0) # 행기준 _ 인덱싱이 0 1 2 - 0, 1이라 concat이후에는 꼭 indexing확인 필요
```

```python
pd.concat([df1, df2], axis = 1) # 이렇게 잘못설계해 붙이면 N/A가 발생해 곤란해질 수 있다
```

```python
# 세개 합병
display(pd.concat([df1, df2, df3], axis = 0))
```

```python
display(pd.concat([df1, df2, df3], axis = 1)) # NAN값에 유의해야한다.
```

```python
## 인덱스 재정렬 _ ignore_index = True 해주기
display(pd.concat([df1, df2, df3], axis = 1, ignore_index = True))
```

### Merge

- Concat과의 차이점: join이 가능하다
  - sql, pk, 공통의 컬럼으로 데이터를 병합할 수 있다
- 공통의 컬럼 기준이 필요할 것
- 그 컬럼의 기준을 통해서 데이터를 병합하기
- **처음 작성한 데이터프레임이 기준이 된다** (= 첫번재가 된다)
  - ignore_index 가 필요없다

- left - 병합한 첫번째 프레임
- right : 병합한 두번째 데이터 프레임
- how: 병합 방식 옵션, default = inner, 기타 left, right, outer이 있다
- on: 공통 컬럼 지정하기. (지정하지 않을 시 알아서 공통열을 사용한다) _ 어지간히 병합할 때 지정해주는 게 좋다
- left_on : 공통열을 지정하는데 두 테이블의 컬럼명이 다르지만 공통인 경우에는 직접 지정해서 컬럼 세팅이 필요하다
- right_on : 비슷하지만 오른쪽으로 붙여주는 역할
- left_index:  인덱스로 병합하기
- right_index: 마찬가지

```python
df1 = pd.DataFrame({'이름':['홍길동','박길동','이길동','최길동','오길동'],
                   '분반':['프로그래밍','데이터분석','데이터사이언스','인공지능','데이터리터러시']})

df2 = pd.DataFrame({'이름':['홍길동','박길동','이길동','최길동','오길동'],
                   '점수':[100,80,90,80,75]})
```

```python
## 기본적인 merge
## 알아서 공통의 컬럼이 이름이기 때문에 잡아준다.
pd.merge(df1,df2)
```

```python
pd.merge(df1, df2, how = 'inner')
```

```python
pd.merge(df1, df2, how = 'left')
```

```python
pd.merge(df1, df2, how = 'outer')
```

```python
pd.merge(df1, df2, how = 'right')
```

## Join

- 알아두면 좋지만 index 등의 제한으로 대부분의 경우 merge가 선호됨
- df간에 중복데이터가 있다면 : 같은 데이터 중복이 생긴다
  - 공통 사용 컬럼 pk에 중복값을 ㅘㄱ인해야함
  - df1에 홍길동, 프로그래밍  100 , df2에 홍길동, 데분, 80 있으면 merge하면 표 개수가 늘어난다
  - on - 병합할 때 사용할 열 이름, 지정하지 않으면 인덱스 기준으로 병합
- how - 동일
- suffix - 컬럼을 지정하라는 뜻"

```python
## 공통 컬럼 join도 2개 이상으로 만들 수 있다.

df1 = pd.DataFrame({'이름':['홍길동','박길동','이길동','최길동','오길동'],
                    '학번':[1,2,3,4,5],
                   '분반':['프로그래밍','데이터분석','데이터사이언스','인공지능','데이터리터러시']})

df2 = pd.DataFrame({'이름':['홍길동','박길동','이길동','유길동','김길동'],
                    '학번':[1,2,3,7,8],
                   '점수':[100,80,90,80,75]})

df3 = pd.DataFrame({'이름':['홍길동','임길동','이길동','유길동','김길동'],
                   '벌점':[5,10,2,6,9]})
```

```python
display(df1, df2)
```

```python
pd.merge(df1,df2,on=['이름','학번']) # on 안에 두 개 이상 Join 컬럼이 필요하면 작성하시면 된다.
```

```python
df2 = pd.DataFrame({'이름':['홍길동','박길동','이길동','유길동','김길동'],
                    '학회원학번':[1,2,3,7,8],
                   '점수':[100,80,90,80,75]})

df3 = pd.DataFrame({'학생이름':['홍길동','임길동','이길동','유길동','김길동'],
                    '분반학번':[1,8,3,7,8],
                    '벌점':[5,10,2,6,9]})
```

```python
display(df2, df3)
```

```python
display(df2, df3)
```

```python
pd.merge(df2, df3, left_on ='이름',right_on='학생이름', how='outer')
```

```python
##index 기준으로 지정할 수 있다.

df2_idx =df2.set_index('이름') # 데이터프레임에서 인덱스를 지정하는 것
df2_idx.reset_index() #인덱스를 다시 푸는 방법
```

```python
#인덱스가 지정된 경우에는 left_index 인자 이용해서 병합할 수 있다.
pd.merge(df2_idx, df3, left_index=True, right_on='학생이름')
```

```python
df2.join(df3) #지정을 하지 않고 바로 join을 하면 인덱스 기준으로 합쳐서 모든 값들이 다 출력이 되었다.
```

```python
#컬럼을 지정하지 않으면 계속해서 인덱스 기준으로 붙여진다.
#df2.join(df3, how='inner', on='이름')


df2 = pd.DataFrame({'이름':['홍길동','박길동','이길동','유길동','김길동'],
                    '학회원학번':[1,2,3,7,8],
                   '점수':[100,80,90,80,75]})

df3 = pd.DataFrame({'이름':['홍길동','임길동','이길동','유길동','김길동'],
                    '분반학번':[1,8,3,7,8],
                    '벌점':[5,10,2,6,9]})
```

```python
# 인덱스를 '이름'으로 설정 후 join 수행
df2_indexed = df2.set_index('이름')
df3_indexed = df3.set_index('이름')

# inner join
result = df2_indexed.join(df3_indexed, how='inner')
print(result)
```

```python
result_outer = df2_indexed.join(df3_indexed, how='outer')
print(result_outer)
# 기타 inner, right, outer 등 가능
```

```python
result_suffixes = df2_indexed.join(df3_indexed, how='inner', lsuffix='_df2', rsuffix='_df3')
print(result_suffixes)
```

### numpy 데이터 병합

- 메모리 효율이 넘파이가 더 좋아 자주 사용된다
- array배열로 1차원, 2차원 간의 연산, 병합 진행 시 주로 사용
- np.concatenate, np.hstack, np.vstack

```python
import numpy as np
```

```python
# concatenate 연산필기
# column_stack

array1 = np.array([[1,2,3],[4,5,6]])
array2 = np.array([[7,8,9],[10,11,12]])
```

```python
np.concatenate((array1, array2), axis=0)
```

```python
np.concatenate((array1, array2), axis=1)
```

```python
np.hstack((array1, array2))
```

```python
np.vstack((array1, array2))
```

```python
##컬럼으로 나눠서 데이터를 병합할 수 있다.
np.column_stack((array1[0], array2[0]))
```
