## 데이터 시각화 (Data Visualization)
- 데이터 시각화를 하는 이유?
- 데이터 분석에 대한 결과물을 더 잘 이해시키고 설득시키기 위한 보조자료
- 근거자료로 역할을 해야 하는 것, 데이터가 왜곡을 주는 그래프를 사용하지 않고 나의 주장을 잘 뒷받침할 수 있는 그래프로 만들어야 하며
- 내가 주장하는 방향성과 그래프가 맥락이 같아야 한다.

----
- 데이터 시각화는 나만의 자료가 아니라 의사소통, 커뮤니케이션에 필요한 자료
- 내가 이해하지 못한 그래프는 사용하면 안 된다.
- 누구나 쉽게 이해할 수 있도록, 데이터가 왜곡되지 않도록, 그 본질을 꼭 잃지 않고 표현하시면 됩니다.

```python
from matplotlib import pyplot as plt
# 세밀한 그래프를 조작 가능
# pyplot 모듈 플롯을 간단하게 생성할 수 있다.
# Figure 도화지를 그린다.
# 그 도화지 안에 작은 그림들을 채우는 형식
# 큰 도화지를 그린 후 -> 세부적인 어떤 그래프를, 그림을  채우는 방식
#%matplotlib inline  #plt.show() 생략해도 되는

plt.figure(figsize=(20,10))
```

```python
# plt 안에 그릴 수 있는 다양한 그래프 함수가 존재
# scatter 등등 그릴 수 있는 함수
# 그 함수들이 그리는 그림이 다르다. 산점도, 막대그래프 등등 그 그래프에 맞춰서 데이터를 잘 정리하고 지정해야 하는 것
#plt.figure(figsize=(20,10))

plt.scatter([1,2,3,4],[4,5,6,7])
plt.title('X and Y')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
## 축에 대한 조정

plt.scatter([1,2,3,4],[4,5,6,7])
plt.xlim(1,2) # 축 조정 세부 조율 가능
plt.title('X and Y')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
## 축에 대한 눈금값 조정하는 방법
## xticks(기존 값, 변경 값) yticks(기존 값, 변경 값)

plt.scatter([1,2,3,4],[4,5,6,7])
plt.xticks([1,2,3,4],['A','B','C','D'])
##plt.xlim(1,2) # 축 조정 세부 조율 가능
plt.title('X and Y')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

### 도화지에 하나의 그림만 그리는 게 아니라 여러 그림을 그리는 경우는
- 현재는 도화지 하나만 그린 상태라서, 그림이 겹쳐서 보여진다.

```python
plt.scatter([1,2,3,4],[4,5,6,7])
plt.scatter([1,2,3,8],[7,15,20,9])
#plt.xticks([1,2,3,4],['A','B','C','D'])
##plt.xlim(1,2) # 축 조정 세부 조율 가능
plt.title('X and Y')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
## 두 개 이상 그래프를 그리려면 범주나 구분자 필요하다.

plt.scatter([1,2,3,4],[4,5,6,7], label= 'apple')
plt.scatter([1,2,3,8],[7,15,20,9], label= 'banana')
#plt.xticks([1,2,3,4],['A','B','C','D'])
##plt.xlim(1,2) # 축 조정 세부 조율 가능
plt.title('X and Y')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend() #범례를 추가하는 함수
```

```python
## 범례 위치도 다양하게 바꿀 수 있다.


plt.scatter([1,2,3,4],[4,5,6,7], label= 'apple')
plt.scatter([1,2,3,8],[7,15,20,9], label= 'banana')
#plt.xticks([1,2,3,4],['A','B','C','D'])
##plt.xlim(1,2) # 축 조정 세부 조율 가능
plt.title('X and Y')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc ='center') #범례를 추가하는 함수
```

```python
## 'sex','embarked','class','alive' 이 피처들의 카운팅을 통해 데이터 분포를 확인하고 싶다.

interested = ['sex','embarked','class','alive']

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12,6))

for idx, variable in enumerate(interested):
    ax = axs[idx//2, idx%2]
    sns.countplot(x= variable, hue='who', data=df, ax=ax)

plt.show()
```

```python
### 리스트랑 함수들 딕셔너리로 매핑하여 시각화

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12,6))

plots = [
    {'func': sns.barplot, 'x':'sex', 'y':'survived', 'title':'Survived Rate by sex', 'ax':axs[0,0]},
    {'func': sns.barplot, 'x':'class', 'y':'survived', 'title':'Survived Rate by class', 'ax':axs[0,1]},
    {'func': sns.histplot, 'x':'age',  'title':'Survived Rate by age', 'ax':axs[1,0]},
    {'func': sns.histplot, 'x':'fare', 'title':'Survived Rate by fare', 'ax':axs[1,1]}
]

for plot in plots:
    plot['func'](data=df, x=plot.get('x'), y=plot.get('y',None),ax=plot['ax'])
    plot['ax'].set_title(plot['title'])

plt.tight_layout() #겹치는 것 방지
plt.show()
```
