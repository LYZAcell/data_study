## Groupby

    - 데이터를 집계할 때 사용
    - 데이터를 묶고 내가 원하는 통계치를 바라볼 때 사용한다.
    
    - 남성 4
    - 여성 2
    - 남성 1
    - 여성 7
    - 남성 4
    
    - groupby
        - 기준컬럼을 잡고 (성별)
        - 남성 - 합계 (9), 평균 (3)
        - 여성 - 합계 (9), 평균 (4.5)
    - 데이터프레임.groupby('기준컬럼')
    
    - groupby 와 함께 사용하는 agg함수 - 통계치를 빠르게 확인할 때 사용
    

```python
pc_gr=tt.groupby('pclass')
# pc_gr을 print할 시 아래와 같은 결과만 나옴. 아래와 같이 반복문 돌려야 결과 확인 가능
# <pandas.core.groupby.generic.DataFrameGroupBy object at 0x7ae984581360>
for i,j in pc_gr:
    print(i,j)
```

```python
# i, j 가 잘 이해되지 않아 다시 출력해봄
# 예시 DataFrame
data = {'pclass': [1, 2, 1, 3, 2, 1, 3],
        'name': ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
        'age': [22, 24, 21, 35, 23, 40, 30]}

tt_ex = pd.DataFrame(data)

# groupby
pc_gr_ex = tt_ex.groupby('pclass')

# 그룹을 순회하며 출력
for i, j in pc_gr_ex:
    print(f'그룹 키 (pclass 값): {i}')
    #print(f'그룹 데이터 (DataFrame):\n{j}\n')
```

```python
# i, j 가 잘 이해되지 않아 다시 출력해봄
# 예시 DataFrame
data = {'pclass': [1, 2, 1, 3, 2, 1, 3],
        'name': ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
        'age': [22, 24, 21, 35, 23, 40, 30]}

tt_ex = pd.DataFrame(data)

# groupby
pc_gr_ex = tt_ex.groupby('pclass')

# 그룹을 순회하며 출력
for i, j in pc_gr_ex:
    print(f'그룹 키 (pclass 값): {i}')
    print(f'그룹 데이터 (DataFrame):\n{j}\n')
```

###  agg
-  sum, mean, median...
- agg(새로운 컬럼명=(기준컬럼, 통계치))...n증가

```python
##groupby를 한 후에 어떤 특정 통계치를 보고 싶은 경우
tt.groupby('pclass').agg(fare_sum=('fare','sum'))
```

```python
##na값은 제외하고 계산이 된다.
tt.groupby('pclass').agg(fare_sum=('fare','sum'),
                        fare_mean=('fare','mean'),
                        age_mean=('age','mean'),
                        age_cnt=('age','count'))
```

```python
##특정 함수를 만들어서 적용 가능하다.
def min_max(x):
    return x.max()- x.min()
```

```python
# agg로 min_max 적용
tt[['pclass','fare','age']].groupby('pclass').agg(min_max)
```

## qcut, cut
- **qcut : 데이터를 분위수(Quantitle) 따라 구간을 나눈다.**
    - 즉, 개수 기준임!!
    - 분위수 기준으로 균등하게 데이터를 나누는 함수
    - pd.qcut(x,q, labels= ??, retbins=??, precision=?, duplicates=??)
    - x는 구간 데이터나눌 시리즈 형태 또는 배열
    - q 구간 나눌 분위수 e.g 4면 4 분위수
    - labels 각 구간에 붙일 이름
    - retbins 구간의 경계를 함께 반환할지?
    - precision 구간의 경계의 소숫점 자릿수


- **cut : 고정된 값의 구간으로 나누는 함수, 데이터의 값 범위에 따라 일정한 크기의 구간을 정의하고, 각 데이터가 해당하는 구간을 속하게 만들면 된다.**
    - 사용자가 구간설정 가능, 숫자 범위기준임
    - pd.cut(x,bins,right=???,나머지 동일, include_lowest=가장 낮은 구간 포함할지 여부?)
    - 위와 동일한 하이퍼파라미터

```python
## qcut 예시 데이터
import numpy as np

data =np.random.randn(10)
print(data)
```

```python
# data , q = 4(4분위로 나누겠다. 5면 5분위)
res_qcut=pd.qcut(data, q=4, labels=['1사분위','2사분위','3사분위','4사분위'])
print(res_qcut)
```

```python
## cut의 예시
data =np.random.randn(10) * 100
print(data)
```

```python
res_cut=pd.cut(data, bins=4, labels=['낮음','중간','높음','매우높음'])
print(res_cut)
```

```python
## 타이타닉 데이터를 예시로 데이터를 나누자!

## 1. age 기준으로 도메인으로 나누기
## na 값은 모른다 하고
## 19미만은 child
## 19~70 Adult
## 나머지는 Senior

## 2. fare 구간을 나누자
## - qcut, cut 나누기

## 구간화를 범주로 만들어서 파생변수를 만들고 확인해 보자!

## 데이터프레임에 특정 함수를 전체에 적용시킬 때 apply()
## apply(내가 전처리할 함수)

## age에 대한 함수 만들기

def age_cat(age):
    if pd.isnull(age):
        return 'Unknown'
    elif age < 19:
        return 'Child'
    elif 19<= age <70:
        return 'Adult'
    else:
        return 'Senior'
```

```python
#apply함수를 이용해서 쉽게 전처리를 할 수 있다.
tt['age_group']=tt['age'].apply(age_cat)
```

```python
##fare 요금에 대한 전처리
tt['fare_quct'] = pd.qcut(tt['fare'], q=4, labels=['1사분위','2사분위','3사분위','4사분위'])
tt['fare_cut']= pd.cut(tt['fare'], bins=4, labels=['1사분위','2사분위','3사분위','4사분위'])
```

```python
tt.fare_quct.value_counts()
```

```python
tt.fare_cut.value_counts()
```

### 데이터 병합
- 붙이는 것
    - 왜 붙일까?
    - 기업에서 데이터는 다 DB 안에 테이블로 나눠서 보관
    - ML 전처리 데이터 병합

- concat
    - 덩어리와 덩어리를 붙이는 것
    - 데이터프레임 + 데이터프레임을 그대로 붙이는 것
    - df1, df2
    - 그대로 붙이는 방법이 2가지가 있다.
        - 행 - 가로로 길어진다.
        - 열 - 세로로 길어진다
- merge
    - 공통된 컬럼을 가지고 병합을 하는 것
    - join - 공통 컬럼 기준을 지정한 후 inner, left, right, outer 조인의 연산 방법
    
        - join - 병합시간에 배울 예정

```python
test1 = pd.DataFrame({'학회원분반번호':[1,2,3,4,5],
                     '학회원분반':['Python','R','SQL','데이터전처리','데이터모델링']})

test2 = pd.DataFrame({'학회원분반번호':[1,2,3,4,6],
                     '학회원수':[100,200,300,400,500]})
```

```python
display(test1)
display(test2)
```

```python
pd.merge(test1, test2, how='inner',on='학회원분반번호')
```

```python
pd.concat([test1,test2],axis=0)
```

```python
pd.concat([test1,test2],axis=1)
```
