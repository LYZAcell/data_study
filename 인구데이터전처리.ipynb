# process
1. 연령별 컬럼을 8~19세로 필터링
  - 만나이로 등록되어있으니 6세~18세?
2. 구에 따라 1,2로 merging
3. 컬럼반전 2008 - 인원으로

```python
# 열 이름 필터링
cols_age = ['연령별','계', '6세','7세','8세','9세','10세','11세','12세','13세','14세','15세','16세','17세','18세']

# 필터링 후 인덱스 리스트로 만들기
drop_index = df[~df['연령별'].isin(cols_age)].index

# 해당 인덱스 드롭하기
df_filtered = df.drop(drop_index)

df_filtered
```

```python
age_filter = df_filtered
```

```python
# 2. 전체 아동수 계산: 연령별로 데이터를 합산
age_total = age_filter.groupby('연도').sum()

# 3. 아동1 (강원, 충청, 전라, 경상) 데이터 계산
regions_아동1 = ['강원', '충청', '전라', '경상']
아동1 = age_filter[age_filter['행정구역(시군구)별'].isin(regions_아동1)].groupby('연도').sum()

# 4. 아동2 데이터 계산 (전체 - 아동1)
아동2 = age_total - 아동1

# 5. 새로운 데이터프레임 생성
df_result = pd.DataFrame({
    '연도': age_filter['연도'],
    '전체아동수': age_total,
    '아동1': 아동1,
    '아동2': 아동2
})

# 결과 확인
print(df_result)
```
