# 1. df transpose전 (실행 쭉쭉)

```python
df = pd.read_csv('./행정구역_연도별_총주민등록인구.csv', encoding='euc-kr')
df.head()
```

```python
# 열 이름 필터링
cols_age = ['연령별', '6세','7세','8세','9세','10세','11세','12세','13세','14세','15세','16세','17세']

# 필터링 후 인덱스 리스트로 만들기
drop_index = df[~df['연령별'].isin(cols_age)].index

# 해당 인덱스 드롭하기
df_filtered = df.drop(drop_index)

df_filtered
```

```python
age_filter = df_filtered.copy()
# '총인구수[명]' 항목만 선택
age_filter = age_filter[age_filter['항목'] == '총인구수[명]']

# 'Unnamed: 21' 열 제거
age_filter = age_filter.drop(columns=['Unnamed: 21'])
# 불필요한 컬럼제거
age_filter = age_filter.drop(columns=['단위','항목'])
age_filter.head(20)
```

```python
# '행정구역(시군구)별'을 인덱스로 설정합니다.
age_filter_seoul = age_filter.set_index('행정구역(시군구)별')

# 이제 '서울특별시' 행을 가져올 수 있습니다.
age_filter_seoul.loc['서울특별시']
```

```python
# '행정구역(시군구)별' 열의 첫 번째 행의 값을 '전체국민'으로 변경
#age_filter.at[0, '행정구역(시군구)별'] = '전체국민'
#age_filter.loc['서울특별시']
```

```python
age_filter
```

```python
age_filter = age_filter.groupby('행정구역(시군구)별').sum()
age_filter = age_filter.drop(columns=['연령별'])
age_filter
```

# 3.지역별 groupby완료
- 아동1, 2 분리
-  컬럼맞추기

```python
#복사
region_df = age_filter.copy()
region_df
```

```python
region_df = region_df.transpose()
region_df
```

```python
region_df.columns
```

```python
# 인덱스의 name 속성 없애기
region_df.index = region_df.index.rename(None)
region_df
```

```python
# 인덱스 이름 제거
region_df.rename_axis(None, inplace=True)

# 'index' 열의 이름을 '연도'로 변경
region_df.rename(columns={'index': '연도'}, inplace=True)

region_df
```

```python
# 첫 행을 연도로 설정
region_df.columns.name = None
region_df = region_df.rename_axis('연도').reset_index()
region_df
```

```python
region_df.columns
region_df.index
```

```python
# '연도' 컬럼을 인덱스로 설정
region_df.set_index('연도', inplace=True)
region_df
```

```python
# 아동1 지역 리스트
child1_regions = ['강원특별자치도','경상남도','경상북도','전라남도','전북특별자치도','충청남도','충청북도']

# 아동1 컬럼 합치기
region_df['아동1'] = region_df[child1_regions].sum(axis=1)

# 아동2 지역 리스트 (아동1, '전체국민' 제외)
#child2_regions = [region for region in region_df.columns if region not in child1_regions + ['전국']]
# 아동2 컬럼 합치기 전에 숫자형으로 변환
#region_df[child2_regions] = region_df[child2_regions].apply(pd.to_numeric, errors='coerce')
#region_df['아동2'] = region_df[child2_regions].sum(axis=1)

child2_regions = ['경기도', '광주광역시', '대구광역시','대전광역시', '부산광역시','서울특별시','세종특별자치시','울산광역시','인천광역시','제주특별자치도']
region_df['아동2']= region_df[child2_regions].sum(axis=1)


# 필요한 컬럼만 저장
#region_df = region_df[['연도','전국', '아동1', '아동2']]
region_df = region_df[['전국', '아동1', '아동2']]

region_df
```

# 4. index 연도 int형으로 바꿔주기

```python
int_df = region_df.copy()
int_df
```

```python
# 인덱스를 컬럼으로 변환
int_df = int_df.reset_index()

# '연도' 컬럼에서 숫자 부분만 추출하고 정수형으로 변환
int_df['연도'] = int_df['연도'].str.extract(r'(\d{4})').astype(int)

# 변경된 '연도' 컬럼을 다시 인덱스로 설정
int_df = int_df.set_index('연도')
```

# 시계열 분석 해보기

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA

# 데이터 로드
data = pd.read_csv('./인구전처리_2차_아동2정정.csv', encoding = 'cp949')
```

```python
# 2024를 제일 아래로 보내기
temp = data.iloc[0].copy()
data = data.drop(data.index[0])
data = pd.concat([data,pd.DataFrame([temp])],ignore_index=True)
data = data.reset_index(drop=True)

data
```

## 현재 아동2 컬럼 정정완료

```python
# 시계열 플롯
# y축단위 100만
plt.figure(figsize=(8,6))
plt.plot(data['연도'],data['전국'], label='total kids')
plt.plot(data['연도'],data['아동1'], label='kids1')
plt.plot(data['연도'],data['아동2'], label='kids2')
plt.legend()
plt.show()
```

```python
# ARIMA 모델링 예시
model = ARIMA(data['전국'], order=(1, 1, 1))
model_fit = model.fit()
print(model_fit.summary())

# 예측
forecast = model_fit.forecast(steps=5)
print(forecast)
```
