## 주소마다 구만 추출하기

````python
import pandas as pd
import json

df = pd.read_json('hospital_list.json')
df
````

````python
# 주소에서 구 정보를 추출하는 함수
def extract_gu(address):
    if isinstance(address, str):
        split_address = address.split()
        for part in split_address:
            if part.endswith('구'):
                return part
    return None

# 구 정보를 추출하여 새로운 열로 추가
df['구'] = df['주소'].apply(extract_gu)
df
````

````python
#컬럼순서 맞춰주기
cols = ['병원명', '주소', '구','예약가능여부', '진료시작시각', '방문자리뷰', '블로그리뷰', '산부인과전문의수', '기타전문의여부']
df = df[cols]
df
````

````python
# 데이터프레임을 JSON 형식으로 변환 (각 행을 JSON 객체로 변환)
json_data = df.to_json(orient='records', force_ascii=False)

# JSON 데이터를 파일로 저장
with open('hospital_list.json', 'w', encoding='utf-8') as f:
    f.write(json_data)

# 코랩에서 파일을 다운로드 링크로 제공
from google.colab import files
files.download('hospital_list.json')
````

````python
# 컬럼순서배치
df = df_ren[['id', 'hospital_name', 'address', 'gu', 'reservation', 'visitcnt', 'blogcnt', 'maindoctorcnt']]
df
````

````python
# 데이터프레임을 JSON 형식으로 변환 (각 행을 JSON 객체로 변환)
json_data = df.to_json(orient='records', force_ascii=False)

# JSON 데이터를 파일로 저장
with open('hospital_fix.json', 'w', encoding='utf-8') as f:
    f.write(json_data)

# 코랩에서 파일을 다운로드 링크로 제공
from google.colab import files
files.download('hospital_fix.json')
````
